```python
from scipy.integrate import simps, quad


x = np.linspace(-3.0, 12.0, 300)
R = 0.1
sigma = 1.0  #conductance
true_csd = 1.0*np.exp(-(x-2.)**2 / (2*np.pi*0.5)) + 0.5*np.exp(-(x-7)**2 / (2*np.pi*1.0))

#print true_csd

fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True)
ax1.plot(x, true_csd)
ax1.set_title('True CSD')

def calculate_pot(csd, z, z0):
    return 1.0/(2 *sigma) * np.trapz( (np.sqrt((z-z0)**2 + R**2 ) - np.abs(z-z0))*csd, z)

pots = [calculate_pot(true_csd, x, x0) for x0 in x]
ax2.plot(x, pots)
ax2.set_title('Potentials')
```

```python
#elec_pos = np.linspace(-3.0, 13.0, 30)
elec_pos = [ -2.0, -1.0, 0.0, 2.2, 4.3, 7.3, 9.0]

meas_pot = np.array([calculate_pot(true_csd, x, x0) for x0 in elec_pos])

plot(x, pots)
scatter(elec_pos, meas_pot)
```

```python
simple_csd = np.array([ -(meas_pot[i-1]+meas_pot[i+1] -2*meas_pot[i])/(elec_pos[i]-elec_pos[i-1])**2 for i in xrange(1, len(elec_pos)-1)])

fig = plt.subplot(1,1,1)
plot(elec_pos[1:len(elec_pos)-1], simple_csd, label="approximated csd")
plot(x, true_csd, label="true csd")
fig.legend(loc=(0.01,0.7))
```

## kCSD workflow

```python
n_sources = 100

source_space = linspace(min(elec_pos), max(elec_pos), n_sources)
#source_space = linspace(-4.4, 4.4, n_sources)
```

```python
#define the positions of the basis functions
xmax=15.0
xmin=-3.0
estimation_area = linspace(xmin, xmax, n_sources)
max_dist = xmax-xmin

dist_density = 300
h = 1.3*(elec_pos[1]-elec_pos[0])
R = 0.15
```

```python
def pot_intarg(src, arg, current_pos, h, R, sigma, src_type):
    """
    """
    if src_type == "gaussian":
        y = (1./(2*sigma))*(sqrt((arg-current_pos)**2 + R**2) - abs(arg - current_pos)) * gauss_rescale(src, current_pos, h)
        # for this formula look at formula (8) from Pettersen et al., 2006
    if src_type == "step":
        #TODO re-evaluate
        y = (1./(2*sigma))*(sqrt((arg-current_pos)**2+R**2) - abs(arg-current_pos)) * (abs(src-current_pos)<0.5*h)
    if src_type == "wavelet":
        pass
            
    return y
```

```python
def gauss_rescale(x, mu, three_stdev):
    """
    returns normalized gaussian scale function centered in mu, cut off farther than three_stdev from the center
    """
    variance = (three_stdev/3.0)**2
    g = 1./sqrt(2. * pi * variance) * exp( -(1./(2.*variance)) * (x-mu)**2 ) #* (abs(x-mu)<three_stdev)
    return g
```

```python
def b_pot_quad(src, arg, h, R, sigma, src_type):
    """
    """
    if src_type in ["gaussian", "step"]:
        x = linspace(src - 4*h, src + 4*h, 150) #manipulate the resolution
        y = [pot_intarg(src, arg, current_pos, h, R, sigma, src_type) for current_pos in x]
        plot(x,y)
        z = np.trapz(y, x)
    else:
        raise error("Source type not implemented!")

    return z
```

```python
def create_dist_table(dist_dens, dist_max, h, R, sigma, src_type):
    """
    Creates table of source contribution to overall potential as a function of distance
    """
    dist_table = zeros(dist_dens+1)

    for i in xrange(0, dist_dens+1):
        dist_table[i] = b_pot_quad(0,(float(i)/dist_dens) * dist_max, h, R, sigma, src_type)
    
    return dist_table

dist_table = create_dist_table(dist_density, max_dist, h, R, sigma, src_type = "gaussian")
#print dist_table
figure()
plot(dist_table)
```

```python
def calculate_b_pot_matrix(x, src, elec_pos, dist_table, dist_dens):
    """ 
    compute the matrix of potentials generated by every source basis function
    at every electrode position. This procedure can be optimized
    bPotMatrix(j,k) = b_j(x_k) 
    with x_k electrode position, b_j j-th basis source from the paper
    """
    n_elec = len(elec_pos)
    
    b_pot_matrix = zeros((n_sources, n_elec))
    
    for src_ind, current_src in enumerate(src):
        for arg_ind, arg in enumerate(elec_pos):
            r = abs(current_src - arg)
            
            #dist_table_ind = uint16(0.85 * r * float(dist_dens)/max_dist)
            dist_table_ind = uint16(r * float(dist_dens)/max_dist)
            #print dist_table_ind
            b_pot_matrix[src_ind, arg_ind] = dist_table[dist_table_ind]
            
    return b_pot_matrix
  
b_pot_matrix = calculate_b_pot_matrix( estimation_area, source_space, elec_pos, dist_table, dist_density)
#print bpm
imshow(b_pot_matrix, interpolation="none")
colorbar()
```

```python
k_pot = dot(transpose(b_pot_matrix), b_pot_matrix) # Kernel matrix K from (2.14) from the paper
```

```python
imshow(k_pot, interpolation="none")
colorbar()
```

```python
def calculate_b_src_matrix(x, src, h):
    n_gx = len(x)

    b_src_matrix = zeros((n_gx, n_sources))

    for src_ind, curr_src in enumerate(src):
        b_src_matrix[:, src_ind] = gauss_rescale(x, curr_src, h)

    return b_src_matrix
    
b_src_matrix = calculate_b_src_matrix( estimation_area, source_space, h)

imshow(b_src_matrix, interpolation="none")
colorbar()
```

```python
interp_cross = dot(b_src_matrix, b_pot_matrix) # cross-kernel matrix from (2.17) in the paper
imshow(interp_cross, interpolation="none")
colorbar()
```

```python
def calculate_b_interp_pot_matrix(x, src, dist_table, dist_dens):
    """
    compute the matrix of potentials generated by every source basis function
    at every electrode position. This procedure can be optimized
    bPotMatrix(j,k) = b_j(x_k) 
    with x_k electrode position, b_j j-th basis source from the paper
    """
    n_points = len(x)
    dist_max = max(x) - min(x)

    b_interp_pot_matrix = np.zeros((n_sources, n_points))
    
    for src_ind, current_src in enumerate(src):
        for arg_ind, arg in enumerate(x):
            r = abs(current_src - arg);
        
            #dist_table_ind = uint16(0.85*r*float(dist_dens)/dist_max)   
            dist_table_ind = uint16(r*float(dist_dens)/dist_max)
            b_interp_pot_matrix[src_ind, arg_ind] = dist_table[dist_table_ind]
    
    return b_interp_pot_matrix

b_interp_pot_matrix = calculate_b_interp_pot_matrix( estimation_area, source_space, dist_table, dist_density)

imshow(b_interp_pot_matrix, interpolation="none")
colorbar()
```

```python
interp_pot = dot(transpose(b_interp_pot_matrix), b_pot_matrix)
imshow(interp_pot, interpolation="none")
colorbar()
```

```python
from sklearn.cross_validation import KFold, LeaveOneOut,ShuffleSplit

lambd=0.0

def calc_err(lambd, pot, k_pot, ind_test, ind_train):
    k_train = k_pot[ind_train, ind_train]
    
    #print "indtest",ind_test
    #print "indtrain", ind_train
    pot_train = pot[ind_train]
    pot_test = pot[ind_test]
    
    beta = np.dot(np.linalg.inv(k_train + lambd*identity(k_train.shape[0])) , pot_train)
    #print "beta.sh", beta.shape
    
    #plot(beta)
    
    #k_cross = np.array([[k_pot[i,j] for i in ind_test] for j in ind_train])
    #k_cross = k_pot[np.array([ind_test, ind_train])]
    k_cross = k_pot[ind_test][:,ind_train]
    #imshow(k_cross, interpolation="none")
    
    pot_est = np.zeros(size(transpose(ind_test)))
    
    pot_est = dot(k_cross, beta)
    
    #plot(pot_est, pot_test)
    print "pottest:", pot_test, "pot_est:", pot_est
    err = linalg.norm(pot_test - pot_est)
    
    return err

#print calc_err(0.0, estimate_pots(), k_pot, np.array([range(0,len(elec_pos)-2)]), np.array([len(elec_pos)-2]))


def cross_validation(lambd, pot, k_pot, n_folds):
    errors = []
    n = len(elec_pos)
    #print "n:",n
    kf = KFold(n, n_folds=n_folds, indices=True)
    #kf = LeaveOneOut(n, indices=True)
    #kf = ShuffleSplit(5, n_iter=15, test_size=0.25, indices=True)
    for ind_train, ind_test in kf:
        #ind_test, ind_train = np.random.permutation(ind_test), np.random.permutation(ind_train)
        err = calc_err(lambd, pot, k_pot, ind_test, ind_train)
        #print err
        errors.append(err)
        
        #print("%s %s" % (ind_train, ind_test))
        
    error = mean(errors)
    print "l=", lambd, ", err=", error
    return error

def choose_lambda(lambdas, n_folds, n_iter):
    """finds the optimal regularization parameter lambda for Tikhonov regularization"""
    n = len(lambdas)
    errors = np.zeros(n)
    errors_iter = np.zeros(n_iter)
    for i, lambd in enumerate(lambdas):
        for j in xrange(n_iter):
            errors_iter[j] = cross_validation(lambd, meas_pot, k_pot, n_folds)
            #errors_iter[j] = calc_cv_error(lambd, n_folds)
        errors[i] = mean(errors_iter)
    fig = pyplot.figure()
    ax = fig.add_subplot(1,1,1)
    plot(lambdas, errors)
    ax.set_xscale('log')
    ax.set_yscale('log')
    return lambdas[errors == min(errors)]

lambdas = np.array([10./j**2 for j in xrange(1,60) ])
#lambdas = np.array([0,1])
lambd = choose_lambda(lambdas, n_folds = 5,  n_iter=1)[0]
#lambd = 0.0
print "lambda =", lambd
```

```python
estimation_table = interp_pot
#lambd=0.0

k_inv = linalg.inv(k_pot + lambd * identity(k_pot.shape[0]))
imshow(k_inv, interpolation="none")
colorbar()
```

```python
def estimate_pots():
    estimation_table = interp_pot
    
    k_inv = linalg.inv(k_pot + lambd *identity(k_pot.shape[0]))
    beta = dot(k_inv, meas_pot)                 
    output = dot(estimation_table, beta)
    return output

output = estimate_pots()
fig, ax = plt.subplots()


output_area = linspace(min(estimation_area), max(estimation_area), len(output))
ax.plot(x, pots, label = "true pot")
ax.plot(elec_pos, meas_pot, label = "interpolated measurements")
ax.plot(output_area, output*max(pots)/max(output), label = "kCSD estimated pot")
#ax.set_ylim([min(pots) ,max(pots)])
ax.legend(loc=(0.01,0.7))
```
